# Engine HP V/s Price

```python
plt.figure(figsize=(10,5))
sns.regplot(x = 'Engine HP', y = 'Price',data = df)
plt.xlabel("Engine HP")
plt.ylabel('Price')
plt.title("Engine HP V/S Price")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x = df['Transmission Type'], y = df['Price'],data = df)
plt.ylabel('Price')
plt.title('Effect of Transmission Type on Price')
plt.show()
```

```python
df.groupby('Transmission Type')['Price'].mean()
```

```python
plt.figure(figsize=(50,20))
sns.barplot(x = df['Make'], y = df['Price'],data=df)
plt.ylabel('Price')
plt.title('Average Price of Each Model')
plt.show()
```

```python
plt.figure(figsize=(29,10))
sns.barplot(x = df['Vehicle Style'],y = df['Price'],data= df)
plt.ylabel('Price')
plt.title('Effect of Vehicle Style on Price')
plt.show()
```

```python
plt.figure(figsize=(29,10))
sns.countplot(x = df['Vehicle Style'], data =df)
```
